This analysis contains 5 trails of data recorded from neurosky headset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot

data = pd.read_csv('Data.csv', index_col=0)
```

```python
sortdata = data.sort_values('size')
sortdata.dropna()
sortdata.head()
```

```python
sortdata = data.sort_values('size')
sd = sortdata.groupby('size', as_index=False)['theta'].mean()

x = data['size']
y = data['theta']
x1 = sd['size']
y1 = sd['theta']

matplotlib.pyplot.title('Size vs Theta wave')
matplotlib.pyplot.scatter(x,y,c='c')
matplotlib.pyplot.scatter(x1,y1,c='y')
matplotlib.pyplot.xlabel('Size')
matplotlib.pyplot.ylabel('Theta Wave')
matplotlib.pyplot.show()
```
